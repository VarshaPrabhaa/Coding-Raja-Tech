```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.utils import resample
major_class = data[(data['Class']==0)] 
minor_class = data[(data['Class']==1)] 

down_sampling = resample(major_class, replace=True, n_samples= 492, random_state=42)  

new_data = pd.concat([down_sampling, minor_class])
```

```python
new_data.shape
```

```python
sns.countplot(new_data,x=new_data['Class'])
```

```python
new_data.groupby('Class').mean()
```

```python
fraud_df = new_data[new_data['Class']==1]
valid_df = new_data[new_data['Class']==0]

fraud_df.describe()
```

```python
valid_df.describe()
```

```python
plt.hist(x=fraud_df.Amount,bins=50)
```

```python
sns.scatterplot(x='Amount',y='Time',data=valid_df)
sns.scatterplot(x='Amount',y='Time',data=fraud_df)
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(new_data.corr(),cmap='Blues',annot=True)
```

```python
from sklearn.model_selection import train_test_split

X = new_data.iloc[:,1:-1]
y = new_data.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=45)
```

```python
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
```

```python
svm.fit(X_train,y_train)
```

```python
svm_predict=svm.predict(X_test)
```

```python
accuracy_score(y_test,svm_predict)
```

```python
 print(classification_report(y_test,svm_predict))
```

```python
confusion_matrix(y_test,svm_predict)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
```

```python
rf.fit(X_train,y_train)
```

```python
rf_predict=rf.predict(X_test)
```

```python
accuracy_score(y_test,rf_predict)
```

```python
print(classification_report(y_test,rf_predict))
```

```python
confusion_matrix(y_test,rf_predict)
```
